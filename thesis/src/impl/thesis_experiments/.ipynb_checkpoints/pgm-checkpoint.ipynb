#### Model description

PGM model s driftem $k$ na $y_t$ má následující podobu 

$$
\begin{alignat}{2}
y_t &= k + \mu_t + \varepsilon_t, \qquad &\varepsilon \sim \mathcal{N}(0, \sigma_{\varepsilon}^2), \\
\mu_t &= \mu_{t-1}^{1+\nu_{t-1}} + \xi_t, \qquad &\xi_t \sim \mathcal{N}(0, \sigma_{\xi}^2), \\
\nu_t &= \rho\nu_{t-1} + \zeta_t, \qquad &\zeta_t \sim \mathcal{N}(0, \sigma_{\zeta}^2),
\end{alignat}
$$

kde $\rho\in[0,1]$ je "discounting" faktor.

Modely tohoto typu jsou populární např. v epidemiologii, kde se uvažuje namísto mocniny přímo exponenciální funkce (a odtud je další cesta k "magickému" reprodukčnímu číslu). Související proces bude logicky velmi citlivý na kombinace hodnot $\nu_t$ a $\mu_t$. Stačí, aby se chviličku udržela nevhodná kombinace a proces exploduje.

Podívejme se na pár realizací. Pro simulaci si nastavíme rozumnou hodnotu $k$, aby nám později filtrace neskončila odmocňováním záporných čísel.

- state evolution model
$$
x_t =
\begin{bmatrix} \mu_t \\ \nu_t \end{bmatrix}
= 
\begin{bmatrix} \mu_{t-1}^{1+\nu_{t-1}} \\ \rho\nu_{t-1} \end{bmatrix}
+ \begin{bmatrix} \xi_t \\ \zeta_t \end{bmatrix},
$$

- measurement model

$$
y_t = H_t x_t + B_t u_t + \varepsilon_t \\
= [1, 0]x_t + k + \varepsilon_t.
$$

#### Initialization

$$
\begin{alignat}{2}
\varepsilon_t &\sim \mathcal{N}(0, 10^2), \\
\xi_t &\sim \mathcal{N}(0, 0.1^2), \\
\zeta_t &\sim \mathcal{N}(0, 0.01^2),
\end{alignat}
$$

The initial parameters are:
- $k=2$,
- $\rho$=0.9,
- $x_0=[100,0]$,
- the length of the series is 500 samples

```python
sigma_epsilon = 10  # sigma_y
sigma_xi = .1
sigma_zeta = .01

k = 2
rho = 0.9

x_0 = [100,0] # [mu_0,nu_0]
ndat = 500


model = power_growth(
    sigma_xi=sigma_xi,
    sigma_zeta=sigma_zeta,
    sigma_epsilon=sigma_epsilon,
    rho=rho,
    k=k,
    ndat=ndat
)
```

#### Definitions of the related matrices

```python
Q = np.diag([sigma_xi**2, sigma_zeta**2])
H = np.array([[1, 0]])
R = np.atleast_2d(sigma_epsilon**2)
```

## Algorithms pre-setup

#### Extended Kalman filter

linearization of the state equation:

$$
F_t =
\begin{bmatrix}
(\nu_{t-1} + 1) \mu_{t-1}^{\nu_{t-1}} &
\mu_{t-1}^{1+\nu_{t-1}} \ln \mu_{t-1} \\
0 & \rho
\end{bmatrix}
$$

```python
class EKF():
    def __init__(self, F, B, H, R, Q, log=False):
        self.F = F 
        self.B = B
        self.H = H
        self.R = R
        self.Q = Q
        self.P = np.eye(self.Q.shape[0]) * 10
        self.x = np.ones(self.Q.shape[0])
        self.log_mean = []                                             # Logger the mean of X var evolution  
        self.rho = 0.9
        if log:
            print(f"EKF filter init")
    
    def predict(self, u=None):              
        mu_t = self.x[0]                    
        nu_t = self.x[1]
        F = np.array([[(nu_t + 1) * mu_t**nu_t, mu_t**(1 + nu_t) * np.log(mu_t)],
                      [0, self.rho]])
        xminus = np.array([mu_t ** (1. + nu_t), rho * nu_t])
        Pminus = F.dot(self.P).dot(F.T) + self.Q
        self.x = xminus
        self.P = Pminus
    
    def correct(self, yt):
        HPHT = self.H.dot(self.P).dot(self.H.T)
        zavorka_inv = np.linalg.inv(HPHT + self.R)
        K = self.P.dot(self.H.T).dot(zavorka_inv)
        innovation = yt - self.H.dot(self.x)
        xplus = self.x + K.dot(innovation)
        zavorka = np.eye(K.shape[0]) - K.dot(self.H)
        KRKT = K.dot(self.R).dot(K.T)
        Pplus = zavorka.dot(self.P).dot(zavorka.T) + KRKT
        self.x = xplus
        self.P = Pplus
        
        # log evoltuin of X var
        self.log_mean.append(self.x)
```

#### PF & ABC filters

```python
# default params
DEFAULT_ALLOWED_RETRIES_CNT = 100
DEFAULT_COUNT_OF_RANDOM_PARTICLES = 5

def state_evolution(self):
    """State evolution equation of SMC filter"""
    particles_prev = self.particles.copy()
    new_particles = []

    indx = 0
    prev_particles_cnt = len(particles_prev)
    allowed_retries_cnt = DEFAULT_ALLOWED_RETRIES_CNT

    while True:
      # obtain current/prev mu, nu
      mu_prev = particles_prev[indx][0]
      nu_prev = particles_prev[indx][1]

      # calculate values of new mu and nu      
      try:
        mu_new = mu_prev**(1 + nu_prev) + norm.rvs(scale=sigma_xi)
        nu_new = nu_prev * rho + norm.rvs(scale=sigma_zeta)

        if (
            mu_new != None and
            not np.iscomplex(mu_new) and
            mu_new >= 0.
        ):
            new_particles.append([
              mu_new,
              nu_new
            ])
            # inc indx/step counter and reset allowed count of retries
            indx += 1
            allowed_retries_cnt = DEFAULT_ALLOWED_RETRIES_CNT
        else:
            allowed_retries_cnt -= 1

            if allowed_retries_cnt < 0:
              indx += 1
              allowed_retries_cnt = DEFAULT_ALLOWED_RETRIES_CNT  

      except ValueError:
        print(f'[DEBUG]: mu_prev={mu_prev}, nu_prev={nu_prev}')

      if indx == prev_particles_cnt:
        break

    remains_cnt = prev_particles_cnt - len(new_particles)

    for indx in range(remains_cnt):
      # take several random particles
      random_particles_cnt = len(new_particles) if len(new_particles) < DEFAULT_COUNT_OF_RANDOM_PARTICLES else DEFAULT_COUNT_OF_RANDOM_PARTICLES
      # generate random particles
      random_particles = random.sample(new_particles, min(len(new_particles), random_particles_cnt))
      new_particles.append([
        sum(np.array(random_particles)[:,0])/len(random_particles),
        sum(np.array(random_particles)[:,1])/len(random_particles)                   
      ])

    # assign newly generated particles
    self.particles = np.array(new_particles.copy())
        
# Assign state evolution function to filters
abcsmc.state_evolution = state_evolution
particlefilter.state_evolution = state_evolution
```

```python
def simulate_y(self):
    """Simulation function for nonlinear example"""
    return np.array(self.particles[:,0]) + k


abcsmc.simulate_y = simulate_y  # Assign respective simulation method to ABC filter
particlefilter.simulate_y = simulate_y
```

#### Init filters

```python
def init_filters():
    # define PF and ABC settings 
    nparticles = 1000                  # No. of particles in filters
    particles_in_ci = 900              # No. of particles to be covered by credibility region
    alpha = .95                        # Credibility region percentage ('p' in paper)
    # Assign particles (samples) from prior
    particles = abs(mvn.rvs(mean=x_0, cov=Q, size=nparticles))
    
    # spawn instances
    # - ekf
    ekf = EKF(F=None, B=2., H=H, R=R, Q=Q)
    # - pf
    pf = particlefilter(Q, nparticles, sigma_epsilon)
    # - abc normal
    abc_normal = abcsmc(Q, nparticles, particles_in_ci, alpha, KernelTypes.Gaussian)
    # - abc cauchy
    abc_cauchy = abcsmc(Q, nparticles, particles_in_ci, alpha, KernelTypes.Cauchy)
    
    
    ekf.rho = rho
    ekf.x = np.array(x_0)
    # assign generated particles
    pf.particles = particles.copy()
    abc_normal.particles = particles.copy()
    abc_cauchy.particles = particles.copy()
    return ekf, pf, abc_normal, abc_cauchy
```

#### SMC propagate func

```python
def propagate(
    model: smc,
    identifier: str,
    step: int,
    y_t: float,
    iteration: int=1
    ):
    if len(model.log_mean) == step:
        mu_prev = model.log_mean[step-1][0]
        if mu_prev >= 0.:
            model.resample()
            model.state_evolution()
            model.smc_update(y_t)
        else:
            print(f'\n[{identifier}] propagation stopped on step t={step}')
```

# Experiment: well-specified

```python
# %%

ITERATIONS = 100

mse_ekf_finals = []
mse_pf_finals = []
mse_abc_normal_finals = []
mse_abc_cauchy_finals = []

print('>> iteration = ')
for it in range(1, ITERATIONS+1):
    print(' {0}'.format(it), end='')

    # generate data
    Y, measurements_noise = model.generate_y(X=X,noisedist='normal')
    initial_measurement = Y[0]
    
    # instantiate filters
    ekf, pf, abc_normal, abc_cauchy = init_filters()
    
    # Do an initial update
    pf.smc_update(initial_measurement)
    abc_normal.smc_update(initial_measurement)
    abc_cauchy.smc_update(initial_measurement)
    ekf.correct(np.atleast_1d(initial_measurement))
    
    # print('>> t = ')
    for t in range(1, ndat):
        # print(' {0}'.format(t), end='')
        # Acquire true measurement y
        y_t = Y[t]

        # Resample and perform state prediction
        propagate(
            model=pf,
            identifier='PF',
            step=t,
            y_t=y_t,
            iteration=it
        )
        propagate(
            model=abc_normal,
            identifier='ABC normal',
            step=t,
            y_t=y_t,
            iteration=it
        )
        propagate(
            model=abc_cauchy,
            identifier='ABC cauchy',
            step=t,
            y_t=y_t,
            iteration=it
        )

        ekf.predict(u=1)
        ekf.correct(np.atleast_1d(y_t))
    
    
    can_pf = True if len(pf.log_mean) == len(X) else False
    can_abc_normal = True if len(abc_normal.log_mean) == len(X) else False
    can_abc_cauchy = True if len(abc_cauchy.log_mean) == len(X) else False

    
    # Calculate MSE
    mse_mu_ekf = mse.calculate(np.array(ekf.log_mean)[:,0], X[:,0])
    mse_nu_ekf = mse.calculate(np.array(ekf.log_mean)[:,1], X[:,1])
    mse_ekf_finals.append([mse_mu_ekf[-1], mse_nu_ekf[-1]])

    if can_pf:
        mse_mu_pf = mse.calculate(np.array(pf.log_mean)[:,0], X[:,0])
        mse_nu_pf = mse.calculate(np.array(pf.log_mean)[:,1], X[:,1])
        mse_pf_finals.append([mse_mu_pf[-1], mse_nu_pf[-1]])

    if can_abc_normal:
        mse_mu_abc_normal = mse.calculate(np.array(abc_normal.log_mean)[:,0], X[:,0])
        mse_nu_abc_normal = mse.calculate(np.array(abc_normal.log_mean)[:,1], X[:,1])
        mse_abc_normal_finals.append([mse_mu_abc_normal[-1], mse_nu_abc_normal[-1]])
    
    if can_abc_cauchy:
        mse_mu_abc_cauchy = mse.calculate(np.array(abc_cauchy.log_mean)[:,0], X[:,0])
        mse_nu_abc_cauchy = mse.calculate(np.array(abc_cauchy.log_mean)[:,1], X[:,1])
        mse_abc_cauchy_finals.append([mse_mu_abc_cauchy[-1], mse_nu_abc_cauchy[-1]])
```

```python
model.save_mse_finals(mse_ekf_finals, 'ekf_normal')
model.save_mse_finals(mse_pf_finals, 'pf_normal')
model.save_mse_finals(mse_abc_normal_finals, 'abc_normal_normal')
model.save_mse_finals(mse_abc_cauchy_finals, 'abc_cauchy_normal')
```
