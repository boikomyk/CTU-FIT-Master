#### Model description

**State** model

$$
\begin{align}
x_t &= A_t x_{t-1} + B_t u_t + w_t, &w_t \sim (0, Q_t),\\
y_t &= h(x_{t}) + \varepsilon_t, &\varepsilon_t \sim (0, R_t).
\end{align}
$$
$$
x_t =
\begin{bmatrix}
x_{1,t} \\ 
x_{2,t} \\ 
v_{x_1,t} \\ 
v_{x_2,t}
\end{bmatrix}
$$

Covariance matrixes

\begin{align*}
    A &=
    \begin{bmatrix}
       1 & 0 & dt & 0 \\
       0 & 1 & 0 & dt \\
       0 & 0 & 1 &  0 \\
       0 & 0 & 0 &  1 
    \end{bmatrix},
    \quad
    &Q &= q^2 \cdot
    \begin{bmatrix}
        \frac{dt^3}{3}    & 0                 & \frac{dt^{2}}{2}  & 0  \\
        0                 & \frac{dt^3}{3}    & 0                 & \frac{dt^{2}}{2} \\
        \frac{dt^{2}}{2}  & 0                 & dt                & 0 \\
        0                 & \frac{dt^{2}}{2}  & 0                 & dt
    \end{bmatrix}
    \notag \\
    R &=
    r^{2}\cdot
    \begin{bmatrix}
        0.01 & 0.0 & 0.0 \\
        0.0 & 0.0001 & 0.0 \\
        0.0 & 0.0 & 0.01
    \end{bmatrix}
\end{align*}
where $dt = 1$, $q = \sqrt{.5}$, $r=3$.

#### Initialization

#### Definitions of the related matrices

```python
# A - update matrix (aka state transition matrix), used to get predicted x
# H - extraction matrix - used to extract the hypothetical measurement if state x is correct and the sensor is perfect
A = np.array([[1, 0, dt, 0],
              [0, 1, 0, dt],
              [0, 0, 1,  0],
              [0, 0, 0,  1]])
H = cartesian_to_polar

# Q - uncertainty covariance matrix of state x (process covariance)
# R - uncertainty covariance matrix of sensor that produces y (measurement covariance)
Q = q**2 * np.array([[dt**3/3, 0      , dt**2/2, 0      ],
                     [0,       dt**3/3, 0,       dt**2/2],
                     [dt**2/2, 0,       dt,      0      ],
                     [0,       dt**2/2, 0,       dt     ]])
R = r**2 * np.array([[0.01, 0.0,    0.0 ],  # 0.09, 0,      0,
                     [0.0,  0.0001, 0.0 ],  # 0,    0.0009, 0
                     [0.0,  0.0,    0.01]]) # 0,    0,      0.09
```

## Algorithms pre-setup

#### Extended Kalman filter

linearization of measurement equation using Jacobian matrix

```python
# linearization of h(x)
def calculate_jacobian(x, THRESH = 0.0001, ZERO_REPLACEMENT = 0.0001):
    """
    Calculates the Jacobian given for four state variables
    Args:
      px, py, vx, vy : floats - four state variables in the system 
      THRESH - minimum value of squared distance to return a non-zero matrix
      ZERO_REPLACEMENT - value to replace zero to avoid division by zero error
    Returns:
      H : the jacobian matrix expressed as a 4 x 4 numpy matrix with float values
    """
    # unwrap
    p_x,p_y,v_x,v_y = x
    
    d_squared = p_x * p_x + p_y * p_y 
    d = math.sqrt(d_squared)
    d_cubed = d_squared * d

    if d_squared < THRESH:
        # print("WARNING: in calculate_jacobian(): d_squared < THRESH")
        H = np.matrix(np.zeros([3, 4]))
    else:
        r11 = p_x / d
        r12 = p_y / d
        r21 = -p_y / d_squared
        r22 = p_x / d_squared
        r31 = p_y * (v_x * p_y - v_y * p_x) / d_cubed
        r32 = p_x * (v_y * p_x - v_x * p_y) / d_cubed

        H = np.array([[r11, r12, 0, 0], 
                      [r21, r22, 0, 0], 
                      [r31, r32, r11, r12]])

    return H
```

```python
# as radar measurement is non-linear, we need to use EKF on Measurement Update cycle,
# but we can use the same formula in prediction cycle.
# we need only linearization for correction step
class EKF():
    def __init__(self, F, B, H, R, Q, log=False):
        self.F = F                          
        self.B = B
        self.H = H
        self.R = R
        self.Q = Q
        self.P = np.eye(self.Q.shape[0]) * 10
        self.x = np.ones(self.Q.shape[0])
        # logger
        self.log_mean = []
        if log:
            print(f"EKF filter init")

    def predict(self, u=None):
        xminus = self.F.dot(self.x)
        Pminus = self.F.dot(self.P).dot(self.F.T) + self.Q
        self.x = xminus
        self.P = Pminus
        
    def correct(self, yt): 
        """
        Args:
        yt : np.array - actual measurement in polar coordinate system
        """
        
        # to convert the non-linear function to linear we use the Taylor series expansion and
        # takes it first derivative. This first order derivative is given by Jacobian Matrix
        # as we are dealing with matrix in the equations
        
        # H   - linearization of h(x)
        # h_x - converts 2d cartesian position and velocity coordinates to polar coordinates 
        
        self.H = calculate_jacobian(self.x) # linearize
        
        HPHT = self.H.dot(self.P).dot(self.H.T)
        zavorka_inv = np.linalg.inv(HPHT + self.R)
        K = self.P.dot(self.H.T).dot(zavorka_inv)
        innovation = yt - cartesian_to_polar(self.x)
        xplus = self.x + K.dot(innovation)
        zavorka = np.eye(K.shape[0]) - K.dot(self.H)
        KRKT = K.dot(self.R).dot(K.T)
        Pplus = zavorka.dot(self.P).dot(zavorka.T) + KRKT
        self.x = np.array(xplus).flatten()
        self.P = Pplus
        
        # log evoltuin of X var
        self.log_mean.append(self.x)
```

#### PF & ABC filters

```python
def state_evolution(self):
    """State evolution equation of SMC filter"""
    particles_prev = self.particles.copy()
    new_particles = []

    for particle_prev in particles_prev:
        x_new = A.dot(np.array(particle_prev)) + mvn.rvs(cov=self.sigma_x)
        new_particles.append(x_new.flatten())

    # assign newly generated particles
    self.particles = np.array(new_particles.copy())

# Assign state evolution function to filters
particlefilter.state_evolution = state_evolution
abcsmc.state_evolution = state_evolution
```

```python
def simulate_y(self):
    """Simulation function for example"""
    simulated_y = []
    for i in range(self.nparticles):
        simulated_y.append(cartesian_to_polar(self.particles[i]))
    
    return np.array(simulated_y)

particlefilter.simulate_y = simulate_y
abcsmc.simulate_y = simulate_y  # Assign respective simulation method to ABC filter
```

#### Init filters

```python
def init_filters():
    # define PF and ABC settings
    nparticles = 1000                  # No. of particles in filters
    particles_in_ci = 900              # No. of particles to be covered by credibility region
    alpha = .95                        # Credibility region percentage ('p' in paper)
    # Assign particles (samples) from prior
    # particles = mvn.rvs(mean=x_0, cov=Q, size=nparticles)
    particles = mvn.rvs(mean=[1,1,1,1], cov=10*np.eye(4), size=nparticles)


    # spawn instances
    # - ekf
    ekf = EKF(F=A, B=None, H=None, R=R, Q=Q)
    # - pf
    pf = particlefilter(Q, nparticles, R)
    # - abc normal
    abc_normal = abcsmc(Q, nparticles, particles_in_ci, alpha, KernelTypes.Gaussian)
    # - abc cauchy
    abc_cauchy = abcsmc(Q, nparticles, particles_in_ci, alpha, KernelTypes.Cauchy)
    
    
    # kf.x = np.array(x_0)
    ekf.x = np.array([1,1,1,1])
    # assign generated particles
    pf.particles = particles.copy()
    abc_normal.particles = particles.copy()
    abc_cauchy.particles = particles.copy()
    return ekf, pf, abc_normal, abc_cauchy
```

#### SMC propagate func

```python
def propagate(
    model: smc,
    identifier: str,
    step: int,
    y_t: float,
    iteration: int=1
    ):
    model.resample()
    model.state_evolution()
    model.smc_update(y_t)
```

# Experiment: well-specified

```python
# %%

ITERATIONS = 100

mse_ekf_finals = []
mse_pf_finals = []
mse_abc_normal_finals = []
mse_abc_cauchy_finals = []

print('>> iteration = ')
for it in range(1, ITERATIONS+1):
    print(' {0}'.format(it), end='')
    
    # generate data
    Y, measurements_noise = model.generate_y(X=X,noisedist='normal')
    initial_measurement = Y[0]
    
    # instantiate filters
    ekf, pf, abc_normal, abc_cauchy = init_filters()
    
    # Do an initial update
    pf.smc_update(initial_measurement)
    abc_normal.smc_update(initial_measurement)
    abc_cauchy.smc_update(initial_measurement)
    ekf.correct(np.atleast_1d(initial_measurement))

    # print('>> t = ')
    for t in range(1, ndat):
        # print(' {0}'.format(t), end='')
        # Acquire true measurement y
        yt = Y[t]

        # Resample and perform state prediction
        propagate(
            model=pf,
            identifier='PF',
            step=t,
            y_t=yt,
            iteration=it
        )
        propagate(
            model=abc_normal,
            identifier='ABC normal',
            step=t,
            y_t=yt,
            iteration=it
        )
        propagate(
            model=abc_cauchy,
            identifier='ABC cauchy',
            step=t,
            y_t=yt,
            iteration=it
        )

        ekf.predict()
        ekf.correct(np.atleast_1d(yt))
    

    # Calculate MSE
    # - x_1 ~ x_1
    mse_x1_ekf = mse.calculate(np.array(ekf.log_mean)[:,0], X[:,0])
    mse_x1_pf = mse.calculate(np.array(pf.log_mean)[:,0], X[:,0])
    mse_x1_abc_normal = mse.calculate(np.array(abc_normal.log_mean)[:,0], X[:,0])
    mse_x1_abc_cauchy = mse.calculate(np.array(abc_cauchy.log_mean)[:,0], X[:,0])

    # - x_2 ~ x_2
    mse_x2_ekf = mse.calculate(np.array(ekf.log_mean)[:,1], X[:,1])
    mse_x2_pf = mse.calculate(np.array(pf.log_mean)[:,1], X[:,1])
    mse_x2_abc_normal = mse.calculate(np.array(abc_normal.log_mean)[:,1], X[:,1])
    mse_x2_abc_cauchy = mse.calculate(np.array(abc_cauchy.log_mean)[:,1], X[:,1])
    
    # - x_3 ~ v_1
    mse_v1_ekf = mse.calculate(np.array(ekf.log_mean)[:,2], X[:,2])
    mse_v1_pf = mse.calculate(np.array(pf.log_mean)[:,2], X[:,2])
    mse_v1_abc_normal = mse.calculate(np.array(abc_normal.log_mean)[:,2], X[:,2])
    mse_v1_abc_cauchy = mse.calculate(np.array(abc_cauchy.log_mean)[:,2], X[:,2])
    
    # - x_4 ~ v_2
    mse_v2_ekf = mse.calculate(np.array(ekf.log_mean)[:,3], X[:,3])
    mse_v2_pf = mse.calculate(np.array(pf.log_mean)[:,3], X[:,3])
    mse_v2_abc_normal = mse.calculate(np.array(abc_normal.log_mean)[:,3], X[:,3])
    mse_v2_abc_cauchy = mse.calculate(np.array(abc_cauchy.log_mean)[:,3], X[:,3])

    # save results
    mse_ekf_finals.append([mse_x1_ekf[-1], mse_x2_ekf[-1], mse_v1_ekf[-1], mse_v2_ekf[-1]])
    mse_pf_finals.append([mse_x1_pf[-1], mse_x2_pf[-1], mse_v1_pf[-1], mse_v2_pf[-1]])
    mse_abc_normal_finals.append([mse_x1_abc_normal[-1], mse_x2_abc_normal[-1], mse_v1_abc_normal[-1], mse_v2_abc_normal[-1]])
    mse_abc_cauchy_finals.append([mse_x1_abc_cauchy[-1], mse_x2_abc_cauchy[-1], mse_v1_abc_cauchy[-1], mse_v2_abc_cauchy[-1]]) 
```

```python
model.save_mse_finals(mse_ekf_finals, 'ekf_normal')
model.save_mse_finals(mse_pf_finals, 'pf_normal')
model.save_mse_finals(mse_abc_normal_finals, 'abc_normal_normal')
model.save_mse_finals(mse_abc_cauchy_finals, 'abc_cauchy_normal')
```

# Experiment: misspecified

```python
# example of measurement with misspecified noise generation
Y_m, measurements_noise_m = model.generate_y(X=X,noisedist='Cauchy')
```

```python
# %%

ITERATIONS = 100

mse_ekf_finals_m = []
mse_pf_finals_m = []
mse_abc_normal_finals_m = []
mse_abc_cauchy_finals_m = []


print('>> iteration = ')
for it in range(1, ITERATIONS+1):
    print(' {0}'.format(it), end='')
    
    # generate data
    Y_m, measurements_noise_m = model.generate_y(X=X,noisedist='Cauchy')
    initial_measurement_m = Y_m[0]
    
    # instantiate filters
    ekf_m, pf_m, abc_normal_m, abc_cauchy_m = init_filters()
    
    # Do an initial update
    pf_m.smc_update(initial_measurement_m)
    abc_normal_m.smc_update(initial_measurement_m)
    abc_cauchy_m.smc_update(initial_measurement_m)
    ekf_m.correct(np.atleast_1d(initial_measurement_m))

    # print('>> t = ')
    for t in range(1, ndat):
        # print(' {0}'.format(t), end='')
        # Acquire true measurement y
        yt = Y_m[t]

        # Resample and perform state prediction
        propagate(
            model=pf_m,
            identifier='PF',
            step=t,
            y_t=yt,
            iteration=it
        )
        propagate(
            model=abc_normal_m,
            identifier='ABC normal',
            step=t,
            y_t=yt,
            iteration=it
        )
        propagate(
            model=abc_cauchy_m,
            identifier='ABC cauchy',
            step=t,
            y_t=yt,
            iteration=it
        )

        ekf_m.predict()
        ekf_m.correct(np.atleast_1d(yt))
    

    # Calculate MSE
    # - x_1 ~ x_1
    mse_x1_ekf = mse.calculate(np.array(ekf_m.log_mean)[:,0], X[:,0])
    mse_x1_pf = mse.calculate(np.array(pf_m.log_mean)[:,0], X[:,0])
    mse_x1_abc_normal = mse.calculate(np.array(abc_normal_m.log_mean)[:,0], X[:,0])
    mse_x1_abc_cauchy = mse.calculate(np.array(abc_cauchy_m.log_mean)[:,0], X[:,0])

    # - x_2 ~ x_2
    mse_x2_ekf = mse.calculate(np.array(ekf_m.log_mean)[:,1], X[:,1])
    mse_x2_pf = mse.calculate(np.array(pf_m.log_mean)[:,1], X[:,1])
    mse_x2_abc_normal = mse.calculate(np.array(abc_normal_m.log_mean)[:,1], X[:,1])
    mse_x2_abc_cauchy = mse.calculate(np.array(abc_cauchy_m.log_mean)[:,1], X[:,1])
    
    # - x_3 ~ v_1
    mse_v1_ekf = mse.calculate(np.array(ekf_m.log_mean)[:,2], X[:,2])
    mse_v1_pf = mse.calculate(np.array(pf_m.log_mean)[:,2], X[:,2])
    mse_v1_abc_normal = mse.calculate(np.array(abc_normal_m.log_mean)[:,2], X[:,2])
    mse_v1_abc_cauchy = mse.calculate(np.array(abc_cauchy_m.log_mean)[:,2], X[:,2])
    
    # - x_4 ~ v_2
    mse_v2_ekf = mse.calculate(np.array(ekf_m.log_mean)[:,3], X[:,3])
    mse_v2_pf = mse.calculate(np.array(pf_m.log_mean)[:,3], X[:,3])
    mse_v2_abc_normal = mse.calculate(np.array(abc_normal_m.log_mean)[:,3], X[:,3])
    mse_v2_abc_cauchy = mse.calculate(np.array(abc_cauchy_m.log_mean)[:,3], X[:,3])

    # save results
    mse_ekf_finals_m.append([mse_x1_ekf[-1], mse_x2_ekf[-1], mse_v1_ekf[-1], mse_v2_ekf[-1]])
    mse_pf_finals_m.append([mse_x1_pf[-1], mse_x2_pf[-1], mse_v1_pf[-1], mse_v2_pf[-1]])
    mse_abc_normal_finals_m.append([mse_x1_abc_normal[-1], mse_x2_abc_normal[-1], mse_v1_abc_normal[-1], mse_v2_abc_normal[-1]])
    mse_abc_cauchy_finals_m.append([mse_x1_abc_cauchy[-1], mse_x2_abc_cauchy[-1], mse_v1_abc_cauchy[-1], mse_v2_abc_cauchy[-1]]) 
```

```python
model.save_mse_finals(mse_ekf_finals_m, 'ekf_cauchy')
model.save_mse_finals(mse_pf_finals_m, 'pf_cauchy')
model.save_mse_finals(mse_abc_normal_finals_m, 'abc_normal_cauchy')
model.save_mse_finals(mse_abc_cauchy_finals_m, 'abc_cauchy_cauchy')
```
